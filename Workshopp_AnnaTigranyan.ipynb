#### 9) Using Basic Assosiation rule display top five movies and score  which shows how much more likely users is to have rated a movie if they have rated Toy Story

```python
movie_association = pd.DataFrame(columns=["movieId","association"])

distinct_movies=np.unique(Ratings["movieId"])


movieID = 1

movie_data=Ratings[Ratings['movieId']==movieID]

for movie1 in distinct_movies:
    
    movie1_data=Ratings[Ratings["movieId"]==movie1]
    
    distinct_cust=len(np.unique(movie_data["userId"]))
    
    distinct_cust1=len(np.unique(movie1_data["userId"]))
    
    movie2_data=movie1_data[movie1_data["userId"].isin(
        np.unique([movie_data["userId"]]))]
    
    distinct_cust_common=len(np.unique(movie2_data["userId"]))
    
    fraction_common_cust=(float( distinct_cust_common)*len(np.unique(Ratings["userId"]))
                         )/(float(distinct_cust)*float(distinct_cust1))
    
    movie_temp=pd.DataFrame(columns=["movieId","association"])
    
    movie_temp = movie_temp.append({
                                    "movieId":  movie1,
                                    "association":  fraction_common_cust
                                    }, ignore_index=True)
    
    movie_association = movie_association.append(movie_temp, ignore_index=True)
```

### Collaborative filtering

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import pairwise_distances
```

#### 10) Create user-user corelation matrix. Find top 5 neighbors for user 23638

```python
Ratings.head()
```

```python
Mean = Ratings.groupby(by="userId",as_index=False)['rating'].mean()

temp = pd.merge(Ratings, Mean, on="userId")
```

```python
temp
```

```python
User_data = pd.pivot_table(temp,values='rating_x',index='userId',columns='movieId')
User_data = User_data.apply(lambda row: row.fillna(row.mean()), axis=1)
User_data.head()
```

```python
b = cosine_similarity(User_data)
np.fill_diagonal(b, 0 )
similarity_with_user = pd.DataFrame(b,index=User_data.index)
similarity_with_user.columns=User_data.index
similarity_with_user.head()
```

```python
def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1)
    return df
```

```python
# top 30 neighbours for each user
sim_user_30_u = find_n_neighbours(similarity_with_user,5)
res = sim_user_30_u.reset_index()

res[res['userId']== 23638]
```

#### 12) Recommend movies to user 51460

```python
TempDF = temp.copy()
TempDF = TempDF.astype({"movieId": str})
Movie_user = TempDF.groupby(by = 'userId')['movieId'].apply(lambda x:','.join(x))
check = pd.pivot_table(temp,values='rating_x',index='userId',columns='movieId')

Movie_data = check.copy()
Movie_data = Movie_data.fillna(Movie_data.mean(axis=0))
```

```python
cosine_s = cosine_similarity(Movie_data)
np.fill_diagonal(cosine_s, 0)

similarity = pd.DataFrame(cosine_s, index=Movie_data.index)
similarity.columns = User_data.index

sim_user_30_m = find_n_neighbours(similarity, 30)

final_movie = Movie_data.copy()
similarity_with_movie = similarity.copy()
```

```python
def User_item_score1(user):
    
    Movie_seen_by_user = check.columns[check[check.index==user].notna().any()].tolist()    
    a = sim_user_30_m[sim_user_30_m.index==user].values
    b = a.squeeze().tolist()
    d = Movie_user[Movie_user.index.isin(b)]
    l = ','.join(d.values)  
    Movie_seen_by_similar_users = l.split(',')
    Movies_under_consideration = list(set(Movie_seen_by_similar_users)-set(list(map(str, Movie_seen_by_user))))
    Movies_under_consideration = list(map(int, Movies_under_consideration))
    score = []
    for item in Movies_under_consideration:
        c = final_movie.loc[:,item]
        d = c[c.index.isin(b)]
        f = d[d.notnull()]
        avg_user = Mean.loc[Mean['userId'] == user,'rating'].values[0]
        index = f.index.values.squeeze().tolist()
        corr = similarity_with_movie.loc[user,index]
        fin = pd.concat([f, corr], axis=1)
        fin.columns = ['adg_score','correlation']
        fin['score']=fin.apply(lambda x:x['adg_score'] * x['correlation'],axis=1)
        nume = fin['score'].sum()
        deno = fin['correlation'].sum()
        final_score = avg_user + (nume/deno)
        score.append(final_score)
    data = pd.DataFrame({'movieId':Movies_under_consideration,'score':score})
    top_5_recommendation = data.sort_values(by='score',ascending=False).head(5)
    Movie_Name = top_5_recommendation.merge(Movies, how='inner', on='movieId')
    Movie_Names = Movie_Name.title.values.tolist()
    return Movie_Names
```
